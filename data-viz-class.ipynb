```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fitness_data.csv")
df.head()
```

```python
plt.plot(df["steps"])
plt.title("Raw Steps Data")
plt.ylabel("Steps")
plt.xlabel("Row of Data")

plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["date"], df["steps"])
plt.title("Number of Steps per Day")
plt.xlabel("Date")
plt.ylabel("Number of steps")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["date"], df["resting_hr"], color="darkred", linewidth = 0.5)
plt.title("Resting Heart Rate Over Time", fontsize=20)
plt.xlabel("Date")
plt.ylabel("Resting HR(bpm)")
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.figure(figsize=(6,6))
plt.title("Steps Versus Calories Burned", fontsize=20)
plt.xlabel("Steps", fontsize=15)
plt.ylabel("Calories Burned",fontsize=15)
plt.scatter(df["steps"],df["calories_burned"], alpha=0.6, color = "purple")
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(7,7))
colors = {"Low":"lightblue", "Moderate":"green","High":"orange"}
for level, subset in df.groupby("activity_level"):
    plt.scatter(subset["steps"], subset["calories_burned"],
    label = level, color = colors[level], alpha = 0.6)
plt.title("Steps Versus Calories Burned by Activity Level", fontsize=20)
plt.xlabel("Steps", fontsize=15)
plt.ylabel("Calories Burned",fontsize=15)
plt.legend()
```

```python
plt.hist(df["steps"],bins = 20, color = 'purple',edgecolor = 'black')
plt.title("Distribution of daily steps")
plt.xlabel("Steps")
plt.ylabel("Frequency of Days")
plt.show()
```

```python
fig, axes = plt.subplots(1,2)
axes[0].hist(df["sleep_hours"], bins =15, color = "blue",edgecolor = "black")
axes[0].set_title("Sleep Hours Distribution")
axes[1].hist(df["resting_hr"], bins =15, color = "darkred", edgecolor = "black")
axes[1].set_title("Resting Heart Rate (bpm) Distribution")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))
# figure is the whole sheet of paper to have two rows, 2/2 array of objects is the axes
#Histogram 1 (top left)
axes[0,0].hist(df["steps"], bins = 20, color = "skyblue",edgecolor = "black")
axes[0,0].set_title("Daily Steps")
#Histogram 2 (top right)
axes[0,1].hist(df["calories_burned"], bins = 20, color = "orange",edgecolor = "black")
axes[0,1].set_title("Calories Burned")
#Histogram 3 (bottom left)
axes[1,0].hist(df["sleep_hours"], bins = 20, color = "green",edgecolor = "black")
axes[1,0].set_title("Number of hours of Sleep")
#Histogram #4(bottom right)
axes[1,1].hist(df["resting_hr"], bins = 20, color = "darkred",edgecolor = "black")
axes[1,1].set_title("Resting Heart Rate")
```

```python
activity_counts = df["activity_level"].value_counts()
```

```python
plt.bar(activity_counts.index,activity_counts.values, color ="brown")
plt.xlabel("Activity Level")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column="sleep_hours", by="activity_level")
plt.title("Sleep Hours By Activity Level")
plt.suptitle("")
plt.xlabel("Activity_level")
plt.ylabel("Number of hours of Sleep")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column="resting_hr", by="activity_level")
plt.title("Resting heart rate By Activity Level")
plt.suptitle("")
plt.xlabel("resting heart rate")
plt.ylabel("Number of hours of Sleep")
plt.show()
```

```python
high_stress = df[df["stress_score"]>60]
# rows of df where stress score >70
high_stress.head()
```

```python
plt.scatter(high_stress["sleep_hours"], high_stress["resting_hr"], color = "navy", alpha = 0.6)
plt.title("Sleep vs Resting Heart Rate--High Stress Days")
plt.xlabel("Sleep Hours")
plt.ylabel("Resting Heart Rate")
plt.xlim(0,10)
plt.ylim(0,70)
plt.show()
```

```python
jan=df[df["date"]< "2024-02-01"]
```

```python
fig, axes = plt.subplots(2,1, figsize=(10,8), sharex=True)

axes[0].plot(jan["date"], jan["resting_hr"], color="darkred")
axes[0].set_title("Resting Heart Rate Over Time")

axes[1].plot(jan["date"], jan["sleep_hours"], color="blue")
axes[1].set_title("Sleep Hours Rate Over Time")

plt.xticks(rotation=45)
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize =(15,5), sharex=True, sharey=True)
# i want to arrange subplots on big viz, one row and three subplots in that row
levels=["Low","Moderate","High"]
colors = ["skyblue","Limegreen","orange"]
# zip_demo=zip(axes,levels,colors)
# print(zip_demo[1])
for ax ,level,color in zip(axes, levels, colors):
    subset = df[df["activity_level"] == level]
    ax.scatter(subset["steps"], subset["calories_burned"], alpha =0.6, color =color )
    ax.set_title(f"{level} Activity")
    ax.set_xlabel("Steps")
    ax.set_ylabel("Calories Burned")
plt.tight_layout()
plt.show()
#
```
