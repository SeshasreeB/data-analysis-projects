```python
students["course_grade"] = students["course_grade"].replace("\"N/A\"", np.nan)
```

```python
students = students.drop_duplicates()
```

```python
#students["student_id"].count()
```

```python
students["first_name"].str.strip().str.upper()
#space in names
```

```python
students["first_name"] = students["first_name"].str.strip().str.upper()
```

```python
students["last_name"] = students["last_name"].str.strip().str.upper()
```

```python
students["major"] = students["major"].str.strip().str.capitalize()
students["major"]= students["major"].replace({
"Comp sci": "Computer Science",
"Cs": "Computer Science",
"Computer science":"Computer Science",
"Bio": "Biology",
"Psych":" Psychology"
})
```

```python
students["major"].value_counts()
```

```python
students.loc["ninety","course_grade"] = 90
```

```python
#students["age"].
```

```python
students["age"].value_counts(dropna = False)
```

```python
students["age"] = pd.to_numeric(students["age"], errors = "coerce")
```

```python
students["age"].describe()
```

```python
students.loc[students["age"] < 10, ["age"]] = np.nan
students.loc[students["age"] >120, ["age"]] = np.nan
```

```python
students.loc[students["age"] < 10, ["age"]] = students["age"].mean()
students.loc[students["age"] >120, ["age"]] = students["age"].mean()
```

```python
students["age"].mean()
```

```python
# create a column combining first and last names
students["full_name"] = students["first_name"] + ' ' + students["last_name"]
students["full_name"]
```

```python
students["age_group"] = pd.cut(
    students["age"],
    bins=[0, 25, 40, 65],
    labels=["Young Adults", "Working adults", "Retired"]
)
```

```python
students[["age","age_group"]].value_counts(dropna=False)
```

```python
students["cost_per_semester"] = students['tuition_paid'] / 3
```
