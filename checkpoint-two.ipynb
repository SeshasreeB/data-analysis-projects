```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.colors as mcolors


# Read data
df = pd.read_csv(r"C:\Users\launchcode\data-analysis-projects\World-happiness-report-updated_2024.csv",encoding='latin1')

# Preview the first few rows to understand the structure
print("Preview of dataset:")
display(df.head())

# print("Summary of DataFrame:")
df.info()

# Descriptive statistics of a DataFrame
print ("Descriptive statistics of a DataFrame:")
df.describe()

# Check shape (rows and columns)
print("\nDataset shape (rows, columns):", df.shape)

# # # List the column names to understand what variables are available
print("\nColumn names:")
print(df.columns.tolist())

# Check data types to see which columns may need cleaning
print("\nData types:")
print(df.dtypes)

# # Check for missing values — important part of EDA
print("\nMissing values per column:")
print(df.isna().sum())

# Check for duplicate rows — helps with data integrity
print("\nNumber of duplicate rows:", df.duplicated().sum())

# # Basic summary of numeric columns
print("\nSummary statistics:")
print(df.describe(include='number').transpose())
```

```python
region_df = happy_df["Regional indicator"].value_counts()
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# This graph shows information about how many countries there are in the regions.
colors = [
    "olive","teal", "yellowgreen", "orange",  "orchid",
    "brown", "sienna", "gray", "green", "purple"
]
plt.figure(figsize=(13,5))
plt.bar(region_df.index, region_df.values, color=colors)
plt.xlabel("Regional Indicator")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter happiest and unhappiest countries
happy_happiest_unhappiest = happy_df[
    (happy_df["Ladder score"] > 7.5) | (happy_df["Ladder score"] < 3.5)
].sort_values(by="Ladder score")


# Create figure
plt.figure(figsize=(10, 8))

# Create horizontal bar chart
plt.barh(
    happy_happiest_unhappiest["Country name"],
    happy_happiest_unhappiest["Ladder score"],
    color='lightblue'

    )

# Labels and title
color = []
plt.title("Happiest and Unhappiest Countries in 2024")
plt.xlabel("Ladder Score")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
