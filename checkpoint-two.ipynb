```python
list_features = [
    'Social support',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption'
]

happy_df[list_features].boxplot(
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor='teal', color='black'),
    medianprops=dict(color='white'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    showmeans = True,
    meanprops={'marker':'o', 'markerfacecolor':'red', 'markeredgecolor':'black'}    )
plt.title('Distribution of Well-being Factors')
plt.xlabel('Score')
plt.show()
```

```python
list_features = [
    'Ladder score',
    'Log GDP per capita'
]

happy_df[list_features].boxplot(
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor='orange', color='black'),
    medianprops=dict(color='blue'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    showmeans = True,
    meanprops={'marker':'o', 'markerfacecolor':'blue', 'markeredgecolor':'black'}
    )
plt.title('Distribution of Well-being Factors')
plt.xlabel('Score')
plt.show()
```

```python
list_features = [
    'Healthy life expectancy'
]

happy_df[list_features].boxplot(
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor='brown', color='black'),
    medianprops=dict(color='blue'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    showmeans = True,
    meanprops={'marker':'o', 'markerfacecolor':'yellow', 'markeredgecolor':'black'}
    )
plt.title('Distribution of Well-being Factors')
plt.xlabel('Score')
plt.show()
```

```python
# Filter happiest and unhappiest countries
happy_happiest_unhappiest = happy_df[
    (happy_df["Ladder score"] > 7.5) | (happy_df["Ladder score"] < 3.5)
].sort_values(by="Ladder score")

#create a color list with 'green' for happiest ,'orange' moderately low happiness and 'red' for unhappies
colors = [
    'green' if score >7.5 else 'orange' if score>3 else  'red'
    for score in happy_happiest_unhappiest["Ladder score"]
]
# Create figure
plt.figure(figsize=(10, 8))

# Create horizontal bar chart
plt.barh(
    happy_happiest_unhappiest["Country name"],
    happy_happiest_unhappiest["Ladder score"],
    color=colors

    )

# Labels and title
color = []
plt.title("Happiest and Unhappiest Countries in 2024")
plt.xlabel("Ladder Score")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
